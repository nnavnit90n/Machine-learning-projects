```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
%matplotlib inline
```

```python
oyo=pd.read_csv('C:\\Users\\Admin\\Desktop\\Skill Circle\\dataset\\oyorooms.csv')
oyo
oyo.dropna(inplace=True)
oyo
```

```python
oyo.describe()
```

```python
oyo=oyo.rename(columns={'OLD PRICE: 4':'old_price'})
oyo
```

```python
o=oyo["OLD PRICE"].str.split(' ',n=1,expand=True)
o
```

```python
o1=oyo["PRICE AFTER DISCOUNT"].str.split(' ',n=1,expand=True)
o1
```

```python
x=o[1].values.reshape(-1,1)
y=o1[1].values.reshape(-1,1)
```

```python
print(len(x))
print(len(y))
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=1/3,random_state=0)
```

```python
from numpy import array
lm=LinearRegression()
lm.fit(x_train,y_train)
# for printing the dofficient
print(lm.coef_)
#for printing intercept value
print(lm.intercept_)
```

```python
ypred=lm.predict(x_test)
print(ypred)
```

```python
accuracy=lm.score(x_test,y_test)
print(accuracy)
```
